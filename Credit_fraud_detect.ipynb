Importing Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

```

Loading Dataset


```python
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
#Shows the first 5 rows of the dataset
credit_card_data.head()
```

```python
#Last 5 rows of the datset
credit_card_data.tail()
```

```python
#Dataset information
credit_card_data.info()
```

```python
#Checking for missing values
credit_card_data.isnull().sum()
```

```python
#Distribution of Legit transaction and Fraud transaction
credit_card_data['Class'].value_counts()
```

We can see the datast is highly unbalanced

0 -> Normal Transaction 

1 -> Fraudulent Transaction

```python
#Seperating teh dataset for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape,fraud.shape)
```

```python
#Statistical analysis of legit data
legit.Amount.describe()
```

```python
#Statistical analysis of Fraud data
fraud.Amount.describe()
```

```python
#Compare the values for both transaction
credit_card_data.groupby('Class').mean()
```

Under Sampling

Build a sample dataset containing similar distribution of Normal transaction and Fraudulent transaction



```python
legit_sample = legit.sample(n=492)
```

Concatenating two Dataframes

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into Features and Target

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

Split the data into training data and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=10)
```

```python
print(X.shape, X_train.shape, X_test.shape)
print(Y.shape, Y_train.shape, Y_test.shape)
```

Model training

Logistic Regression

```python
model = LogisticRegression()
```

```python
#Traing the Logistic Regression model with Training Data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
#Accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accurarcy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on training data: ', training_data_accurarcy)
```

```python
#Accuracy of test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on test data : ', test_data_accuracy)
```
